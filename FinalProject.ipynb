Each document in this corpus corresponds to a single verse from the King James Version (KJV) Bible. The dataset provides the following fields for each verse:

- `id`: a unique numeric identifier for the verse  
- `b`: book number  
- `c`: chapter number

- `t`: the verse text  

These fields directly encode an OHCO (Ordered Hierarchy of Content Objects) structure:

Book → Chapter → Verse → Text

At the highest level, the corpus is divided into 66 books. Each book contains one or more chapters, and each chapter contains one or more verses. In this project, each verse is treated as a document-level unit suitable for modeling and analysis.

The internal structure of each document consists of a single uninterrupted text string written in Early Modern English. While linguistic and rhetorical form varies by genre (e.g., poetic parallelism in Psalms, legal prescriptions in Leviticus, narrative prose in Genesis, epistolary structures in Pauline letters), structurally each document contains only one verse-level text field without paragraph subdivisions.

This consistent hierarchical encoding enables clean transformation into F-compliant tables and supports aggregation and modeling at the verse, chapter, book, or genre levels.
