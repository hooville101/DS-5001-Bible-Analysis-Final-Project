```python
import os
import pandas as pd
```

```python
pca_scores_df = pd.read_csv("../data/output/PCA_DOC_SCORES.csv")

print("PCA document scores shape:", pca_scores_df.shape)
pca_scores_df.head()
```

```python
os.makedirs("../data/output", exist_ok=True)

pca_scores_df.to_csv(
    "../data/output/PCA_DCM.csv",
    index=False
)

print("PCA Document–Component Matrix saved.")
```

```python
doc_index_df = pd.read_csv("../data/output/tfidf_doc_index.csv")

pca_scores_df.insert(0, "Document", doc_index_df.iloc[:, 0])

pca_scores_df.to_csv(
    "../data/output/PCA_DCM_with_docs.csv",
    index=False
)

print("DCM with document labels saved.")
```
