```python
import os
import numpy as np
import pandas as pd
from scipy.sparse import load_npz
from sklearn.decomposition import PCA
```

```python
tfidf_l2 = load_npz("../data/output/TFIDF_L2_REDUCED.npz")

print("TFIDF_L2_REDUCED shape:", tfidf_l2.shape)
```

```python
terms = pd.read_csv(
    "../data/output/tfidf_term_index_reduced.csv"
).squeeze().tolist()

print("Number of reduced terms:", len(terms))
print("Sample terms:", terms[:5])
```

```python
X = tfidf_l2.toarray()

print("Dense matrix shape:", X.shape)
```

```python
n_components = 10

pca = PCA(n_components=n_components, random_state=5001)

components = pca.fit(tfidf_l2.toarray()).components_

print("Components shape:", components.shape)
```

```python
print("Explained variance ratio:")
print(pca.explained_variance_ratio_)

print("\nCumulative variance:")
print(np.cumsum(pca.explained_variance_ratio_))
```

```python
pc1 = components[0]

top_pos_pc1_idx = np.argsort(pc1)[-5:][::-1]
top_pos_pc1_terms = [terms[i] for i in top_pos_pc1_idx]

print("Top 5 positive terms for PC1:")
print(top_pos_pc1_terms)
```

```python
pc2 = components[1]

top_neg_pc2_idx = np.argsort(pc2)[:5]
top_neg_pc2_terms = [terms[i] for i in top_neg_pc2_idx]

print("Top 5 negative terms for PC2:")
print(top_neg_pc2_terms)
```

```python
import plotly.express as px
import os

pc1_vals = components[0]
pc2_vals = components[1]

plot_df = pd.DataFrame({
    "PC1": pc1_vals,
    "PC2": pc2_vals,
    "Word": terms
})

fig = px.scatter(
    plot_df,
    x="PC1",
    y="PC2",
    hover_name="Word",
    title="PCA Loadings: PC1 vs PC2"
)

fig.update_layout(
    width=900,
    height=700
)

save_path = "/Users/nicholasthornton/Downloads/DS 5001/DS-5001-Bible-Analysis-Final-Project/graphs"

os.makedirs(save_path, exist_ok=True)

fig.write_html(os.path.join(save_path, "PC1_vs_PC2_interactive.html"))

print("Interactive PCA plot saved to graphs folder.")

fig
```
