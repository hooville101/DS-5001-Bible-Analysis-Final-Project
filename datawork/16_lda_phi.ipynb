```python
import os
import pandas as pd
import numpy as np
```

```python
delimiter = "|"

lda_loadings_df = pd.read_csv("../data/output/LDA_LOADINGS.csv", index_col=0)
lda_top_words_df = pd.read_csv("../data/output/LDA_TOP_WORDS.csv")

print("Phi matrix shape:", lda_loadings_df.shape)
lda_loadings_df.head()
```

```python
phi = lda_loadings_df.div(lda_loadings_df.sum(axis=1), axis=0)

phi.head()
```

```python
phi.sum(axis=1)
```

```python
n_top = 10

phi_top_words = {}

for topic in phi.index:
    top_terms = (
        phi.loc[topic]
        .sort_values(ascending=False)
        .head(n_top)
    )
    phi_top_words[topic] = top_terms

phi_top_words
```

```python
delimiter = "|"
print("Delimiter:", delimiter)
```

```python
os.makedirs("../data/output", exist_ok=True)

phi.to_csv("../data/output/LDA_PHI_PROBABILITIES.csv")

print("LDA phi probabilities saved.")
```
