```python
import os
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
pca_scores_df = pd.read_csv("../data/output/PCA_DOC_SCORES.csv")

print("PCA Scores shape:", pca_scores_df.shape)
pca_scores_df.head()
```

```python
metadata_df = pd.read_csv("../data/output/tfidf_doc_index.csv")

print(metadata_df.head())
```

```python
doc_labels = metadata_df.iloc[:, 0].tolist()

print("Sample doc labels:", doc_labels[:5])
print("Total documents:", len(doc_labels))
```

```python
doc_plot_df = pca_scores_df.copy()
doc_plot_df["Document"] = doc_labels

# Extract book number
doc_plot_df["BookNum"] = (
    doc_plot_df["Document"]
    .str.split("_")
    .str[0]
    .astype(int)
)

doc_plot_df.head()
```

```python
key_df = pd.read_csv("../data/archive/key_english.csv")

print(key_df.head())
```

```python
book_map = dict(zip(key_df["b"], key_df["n"]))

print(list(book_map.items())[:5])
```

```python
doc_plot_df = pca_scores_df.copy()
doc_plot_df["Document"] = doc_labels

# Extract book number
doc_plot_df["BookNum"] = (
    doc_plot_df["Document"]
    .str.split("_")
    .str[0]
    .astype(int)
)

# Map using key_english.csv
doc_plot_df["BookName"] = doc_plot_df["BookNum"].map(book_map)

# Combined label
doc_plot_df["BookLabel"] = (
    doc_plot_df["BookNum"].astype(str)
    + " - "
    + doc_plot_df["BookName"]
)

doc_plot_df.head()
```

```python
# Map Testament using key_english.csv
testament_map = dict(zip(key_df["b"], key_df["t"]))

doc_plot_df["Testament"] = doc_plot_df["BookNum"].map(testament_map)

doc_plot_df.head()
```

```python
import plotly.graph_objects as go
import numpy as np

fig_docs = go.Figure()

# Ensure books appear in canonical order
books_sorted = (
    doc_plot_df
    .sort_values("BookNum")
    ["BookName"]
    .unique()
)

for book in books_sorted:

    subset = doc_plot_df[doc_plot_df["BookName"] == book]

    subset_ot = subset[subset["Testament"] == "OT"]
    subset_nt = subset[subset["Testament"] == "NT"]

    # Old Testament points (filled)
    if not subset_ot.empty:
        fig_docs.add_trace(
            go.Scatter(
                x=subset_ot["PC1"],
                y=subset_ot["PC2"],
                mode="markers",
                name=book,
                marker=dict(
                    size=5,
                    opacity=0.7
                ),
                text=subset_ot["Document"],
                customdata=np.stack(
                    (subset_ot["BookName"], subset_ot["Testament"]),
                    axis=-1
                ),
                hovertemplate=(
                    "Document: %{text}<br>"
                    "Book: %{customdata[0]}<br>"
                    "Testament: %{customdata[1]}<br>"
                    "PC1: %{x:.3f}<br>"
                    "PC2: %{y:.3f}<extra></extra>"
                )
            )
        )

    # New Testament points (hollow)
    if not subset_nt.empty:
        fig_docs.add_trace(
            go.Scatter(
                x=subset_nt["PC1"],
                y=subset_nt["PC2"],
                mode="markers",
                name=book if subset_ot.empty else None,
                marker=dict(
                    size=5,
                    opacity=0.7,
                    symbol="circle-open"
                ),
                text=subset_nt["Document"],
                customdata=np.stack(
                    (subset_nt["BookName"], subset_nt["Testament"]),
                    axis=-1
                ),
                hovertemplate=(
                    "Document: %{text}<br>"
                    "Book: %{customdata[0]}<br>"
                    "Testament: %{customdata[1]}<br>"
                    "PC1: %{x:.3f}<br>"
                    "PC2: %{y:.3f}<extra></extra>"
                ),
                showlegend=subset_ot.empty
            )
        )

fig_docs.update_layout(
    title="Documents in PC1–PC2 Space<br>Filled = Old Testament, Hollow = New Testament",
    width=1200,
    height=900,
    legend=dict(
        orientation="h",
        yanchor="top",
        y=-0.25,
        xanchor="center",
        x=0.5,
        font=dict(size=9)
    ),
    xaxis_title="PC1",
    yaxis_title="PC2"
)

fig_docs
```

```python
save_path = "/Users/nicholasthornton/Downloads/DS 5001/DS-5001-Bible-Analysis-Final-Project/graphs"

os.makedirs(save_path, exist_ok=True)

fig_docs.write_html(
    os.path.join(save_path, "PCA_documents_PC1_PC2_by_book_and_testament.html")
)

print("Document PCA plot saved.")
```

```python
import plotly.graph_objects as go

# Load PCA loadings
loadings_df = pd.read_csv("../data/output/PCA_LOADINGS.csv", index_col=0)

# Load reduced vocabulary
terms = pd.read_csv(
    "../data/output/tfidf_term_index_reduced.csv"
).squeeze().tolist()

pc1_vals = loadings_df.loc["PC1"].values
pc2_vals = loadings_df.loc["PC2"].values

load_plot_df = pd.DataFrame({
    "PC1": pc1_vals,
    "PC2": pc2_vals,
    "Word": terms
})

fig_load = go.Figure()

fig_load.add_trace(
    go.Scatter(
        x=load_plot_df["PC1"],
        y=load_plot_df["PC2"],
        mode="markers",
        marker=dict(
            size=6,
            opacity=0.7
        ),
        text=load_plot_df["Word"],
        hovertemplate=(
            "Word: %{text}<br>"
            "PC1: %{x:.3f}<br>"
            "PC2: %{y:.3f}<extra></extra>"
        )
    )
)

fig_load.update_layout(
    title="PCA Loadings: PC1 vs PC2",
    width=1000,
    height=800,
    xaxis_title="PC1",
    yaxis_title="PC2"
)

fig_load.add_shape(
    type="line",
    x0=0, x1=0,
    y0=load_plot_df["PC2"].min(),
    y1=load_plot_df["PC2"].max(),
    line=dict(color="black", width=1)
)

fig_load.add_shape(
    type="line",
    y0=0, y1=0,
    x0=load_plot_df["PC1"].min(),
    x1=load_plot_df["PC1"].max(),
    line=dict(color="black", width=1)
)

fig_load
```

```python
fig_load.write_html(
    os.path.join(save_path, "PCA_loadings_PC1_PC2.html")
)

print("Loadings PCA plot saved.")
```

```python
import plotly.graph_objects as go

# First column contains component names
loadings_df = pd.read_csv("../data/output/PCA_LOADINGS.csv")

# Rename first column
loadings_df = loadings_df.rename(columns={loadings_df.columns[0]: "Component"})

# Set component as index
loadings_df = loadings_df.set_index("Component")

# Transpose so rows = terms, columns = PCs
loadings_t = loadings_df.T
loadings_t.index.name = "Term"
loadings_t = loadings_t.reset_index()

loadings_t.head()
```

```python
# Sort by PC1 loading
pc1_sorted = loadings_t.sort_values("PC1")

print("Top 10 negative terms for PC1:")
print(pc1_sorted.head(10)[["Term", "PC1"]])

print("\nTop 10 positive terms for PC1:")
print(pc1_sorted.tail(10)[["Term", "PC1"]])
```

```python
# Sort documents by PC1 score
doc_pc1_sorted = doc_plot_df.sort_values("PC1")

print("Lowest PC1 documents:")
print(doc_pc1_sorted.head(10)[["Document", "BookName", "Testament", "PC1"]])

print("\nHighest PC1 documents:")
print(doc_pc1_sorted.tail(10)[["Document", "BookName", "Testament", "PC1"]])
```

```python
book_pc1_means = (
    doc_plot_df
    .groupby("BookName")["PC1"]
    .mean()
    .sort_values()
)

print(book_pc1_means)
```

```python
testament_pc1_means = (
    doc_plot_df
    .groupby("Testament")["PC1"]
    .mean()
)

print(testament_pc1_means)
```

```python
# Sort documents by PC2 score
doc_pc2_sorted = doc_plot_df.sort_values("PC2")

print("Lowest PC2 documents:")
print(doc_pc2_sorted.head(10)[["Document", "BookName", "Testament", "PC2"]])

print("\nHighest PC2 documents:")
print(doc_pc2_sorted.tail(10)[["Document", "BookName", "Testament", "PC2"]])
```

```python
book_pc2_means = (
    doc_plot_df
    .groupby("BookName")["PC2"]
    .mean()
    .sort_values()
)

print(book_pc2_means)
```

The first principal component reflects a stylistic polarity driven largely by personal and archaic pronoun usage versus more narrative, third-person language. Words with strong positive loadings on PC1 include “I,” “my,” “thy,” “thou,” “thee,” and modal verbs such as “shall” and “will,” indicating passages characterized by direct address and personal speech. In contrast, strong negative loadings include high-frequency narrative terms such as “his,” “son,” “sons,” “was,” “were,” “the,” and “had,” which are more typical of genealogical or historical description. At the document level, books like Psalms, Isaiah, Hosea, and Ruth tend to score highly on PC1, while books such as 1 Chronicles and Nehemiah fall toward the negative side. Because the average PC1 values for the Old and New Testaments are nearly identical, this polarity does not reflect a Testament-level divide but instead distinguishes rhetorical style, separating personal or devotional language from descriptive, third-person narrative structure.
