## Entire Dataset

```python
import os
import glob

data_path = "../data/archive"

all_files = glob.glob(os.path.join(data_path, "*"))

num_raw_documents = len(all_files)
print("Number of raw documents:", num_raw_documents)

total_size_bytes = sum(os.path.getsize(f) for f in all_files if os.path.isfile(f))
total_size_mb = total_size_bytes / (1024 * 1024)
print("Total size of raw documents (MB):", round(total_size_mb, 2))

file_formats = set(os.path.splitext(f)[1] for f in all_files)
print("File formats:", file_formats)
```
