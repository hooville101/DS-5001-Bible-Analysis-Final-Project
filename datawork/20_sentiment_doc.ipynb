```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bow_sent = pd.read_csv("../data/output/BOW_SENT.csv", sep="|")

print("BOW_SENT shape:", bow_sent.shape)
bow_sent.head()
```

```python
emo_cols = [
    "anger", "anticipation", "disgust", "fear",
    "joy", "sadness", "surprise", "trust",
    "positive", "negative"
]
```

```python
doc_sent = (
    bow_sent
    .groupby(["book_number", "chapter", "verse"])[emo_cols]
    .sum()
    .reset_index()
)

doc_sent.head()
```

```python
bow = pd.read_csv("../data/output/BOW.csv", sep="|")

verse_lengths = (
    bow
    .groupby(["book_number", "chapter", "verse"])["n"]
    .sum()
    .reset_index(name="token_count")
)

doc_sent = doc_sent.merge(
    verse_lengths,
    on=["book_number", "chapter", "verse"],
    how="left"
)

for col in emo_cols:
    doc_sent[col] = doc_sent[col] / doc_sent["token_count"]

doc_sent.head()
```

```python
delimiter = "|"
ohco_bag = ["book_number", "chapter", "verse"]

print("Delimiter used:", f'"{delimiter}"')
print("Document bag expressed in OHCO levels:", ohco_bag)
```

```python
output_path = "../data/output/DOC_SENT.csv"

doc_sent.to_csv(
    output_path,
    sep="|",
    index=False
)

print("Saved:", output_path)
```
