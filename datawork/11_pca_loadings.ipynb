```python
import os
import numpy as np
import pandas as pd
from scipy.sparse import load_npz
from sklearn.decomposition import PCA
```

```python
tfidf_l2 = load_npz("../data/output/TFIDF_L2_REDUCED.npz")

print("TFIDF_L2_REDUCED shape:", tfidf_l2.shape)
```

```python
terms = pd.read_csv(
    "../data/output/tfidf_term_index_reduced.csv"
).squeeze().tolist()

print("Number of terms:", len(terms))
print("Sample terms:", terms[:5])
```

```python
n_components = 10

pca = PCA(n_components=n_components, random_state=42)

pca.fit(tfidf_l2.toarray())

components = pca.components_

print("Loadings matrix shape:", components.shape)
```

```python
loadings_df = pd.DataFrame(
    components,
    columns=terms,
    index=[f"PC{i+1}" for i in range(n_components)]
)

loadings_df.head()
```

```python
os.makedirs("../data/output", exist_ok=True)

loadings_df.to_csv(
    "../data/output/PCA_LOADINGS.csv"
)

print("PCA Loadings saved.")
```
